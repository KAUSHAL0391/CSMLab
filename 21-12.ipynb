```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

group1 = np.random.rand(25)
group2 = np.random.rand(20)

variance1 = np.var(group1)
variance2 = np.var(group2)

f_value = variance1 / variance2

df1 = len(group1) - 1
df2 = len(group2) - 1

p_value = stats.f.cdf(f_value, df1, df2)

print('Degree of freedom 1:',df1)
print('Degree of freedom 2:',df2)
print("F-statistic:", f_value)
print("p-value:", p_value)
```

```python
df1 = 7
df2 = 13

sample1 = np.random.chisquare(df1,size=1000)
sample2 = np.random.chisquare(df2,size=1000)

f_value = (sample1/df1)/(sample2/df2)

f_value = np.sort(f_value)
pdf = stats.f.pdf(f_value,df1,df2)
cdf = stats.f.cdf(f_value,df1,df2)

p_value = 1 - cdf

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(f_value,pdf)
plt.title("F-Distribution PDF")
plt.xlabel('x')
plt.grid(True)
plt.ylabel("Probability_density")
plt.subplot(1,2,2)
plt.plot(f_value,cdf)
plt.title("F-dist CDF")
plt.xlabel('x')
plt.ylabel("Cumulative Density")
plt.grid(True)
plt.tight_layout()
plt.show
```

```python
x = np.random.rand(25)
num_group = 5
group_lebel = np.random.randint(0,num_group,size = len(x))
group_mean=[]
for i in range(num_group):
    group_mean.append(np.mean(x[group_lebel==i]))
overall_mean = np.mean(x)

SSB = np.sum([len(x[group_lebel==i])*(group_mean[i]-overall_mean)**2 for i in range(num_group)])
df_between = num_group-1
df_within = len(x) - num_group
MSB = SSB/df_between
SSW = 0
for i in range (num_group):
    group_samples = x[group_lebel==i]
    SSW += np.sum(group_samples - group_mean[i]**2)

MSW = SSW / df_within
f_value = MSB/MSW
print("Degree of Freedom Between Groups",df_between)
print("Degree of Freedom Between Groups",df_within)
```

```python
alpha = 0.005
f_critical = stats.f.ppf(1-alpha,df_between,df_within)
print("f_critical",f_critical)
if f_value > f_critical:
    print("Reject to nulll hypothesis")
else:
    print("fal to reject the null hypothesis")
```
